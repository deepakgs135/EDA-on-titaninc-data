```python
import numpy as np 
import pandas as pd
```

```python
train_data = pd.read_csv(r"D:\internship\prodigy\titanic\titanic (1)\train.csv")
train_data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(train_data.isnull().sum())
```

```python
train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
```

```python
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)
```

```python
train_data.drop('Cabin', axis=1, inplace=True)
```

```python
print(train_data.isnull().sum())
```

```python
sns.countplot(x='Survived', data=train_data)
plt.title('Distribution of Survived')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=train_data)
plt.title('Survival by Pclass')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=train_data)
plt.title('Survival by Sex')
plt.show()
```

```python
sns.histplot(x='Age', hue='Survived', data=train_data, kde=True)
plt.title('Survival by Age')
plt.show()
```

```python
sns.scatterplot(x='Pclass', y='Fare', data=train_data)
plt.title('Relationship between Pclass and Fare')
plt.show()
```
